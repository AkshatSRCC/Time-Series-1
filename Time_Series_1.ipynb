```python
#FROM TEXT TO DATE
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.head()
```

```python
df_comp.date.describe()
```

```python
# Setting The index


df_comp.set_index("date", inplace = True)
df_comp.head()
```

```python
df_comp.date.describe()
# once "date" becomes an index, we no longer save its values as a separate attribute in the data frame
```

```python
# Setting THE DESIRED FREQUENCY
# the values this atrgument can take are specific letters of the alphabet
# 'h' - hourly
# 'w' - weekly
# 'd' - daily
# 'm' - monthly
# 'a' - yearly
df_comp = df_comp.asfreq('d')
df_comp.head()
```

```python
# Handling the missing values
df_comp.isna()
```

```python
df_comp.isna().sum()
```

```python
# "fillna(" method")
# front filling, bACK FILLING, ASSIGNING THE SAME VAlue
df_comp.spx = df_comp.spx.fillna(method = "ffill")
df_comp.isna().sum()
```

```python
df_comp.ftse = df_comp.ftse.fillna(method = "bfill")
df_comp.isna().sum()
```

```python
df_comp.dax = df_comp.dax.fillna(value = df_comp.dax.mean())
df_comp.isna().sum()
```

```python
#Simplifying the Dataset
# we can remove the columns for ftse, dax, nikkei,
# 1) the less data we load , the faster we can manipulate it
# clarity, easier to keep track of the dataset
df_comp['market_value'] = df_comp.spx
df_comp.describe()
```

```python
# Splitting the Data
size = int(len(df_comp)*0.8)
df = df_comp.iloc[:size]
df_test = df_comp.iloc[size:]
df.tail()
```
